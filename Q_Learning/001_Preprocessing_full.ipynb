```python
import pandas as pd
import numpy as np
import talib

# Read NVIDIA stock price data
file_path = "/home/jesse/Projects/RL_Testing/Q_Learning/Training/nvidia_stock_prices.csv"
df = pd.read_csv(file_path)

# Check Date format
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

# Take a look
df.head()
```

```python
# 计Calculate Indexes

# 1.（SMA）
df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)
df['SMA_50'] = talib.SMA(df['Close'], timeperiod=50)

# 2.（EMA）
df['EMA_20'] = talib.EMA(df['Close'], timeperiod=20)
df['EMA_50'] = talib.EMA(df['Close'], timeperiod=50)

# 3.（MACD）
df['MACD'], df['MACD_signal'], df['MACD_hist'] = talib.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)

# 4.（RSI）
df['RSI_14'] = talib.RSI(df['Close'], timeperiod=14)

# 5.（Bollinger Bands）
df['BB_upper'], df['BB_middle'], df['BB_lower'] = talib.BBANDS(df['Close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)

# 6.（ATR）
df['ATR_14'] = talib.ATR(df['High'], df['Low'], df['Close'], timeperiod=14)

# 7.（MFI）
df['MFI_14'] = talib.MFI(df['High'], df['Low'], df['Close'], df['Volume'], timeperiod=14)

# 8.（Stochastic Oscillator）
df['slowk'], df['slowd'] = talib.STOCH(df['High'], df['Low'], df['Close'], fastk_period=14, slowk_period=3, slowd_period=3)

# Take a look
df = df.dropna()  # Drop NaN value
print(df.tail())
```

```python
# Save DataFrame to CSV
output_file_path = "/home/jesse/Projects/RL_Testing/Q_Learning/Training/NVDA_full.csv"
df.to_csv(output_file_path)

print(f"Dataframe saved to : {output_file_path}")
```
